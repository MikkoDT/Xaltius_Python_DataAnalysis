```python
# to close the external all active sources we need to close the file

with open("demo.txt","r") as file:
  print(file.read())
```

```python
# writelines or readlines()
with open("demo.txt","r") as file:
  content = file.readlines()
print(content)
content[1] = "modified data \n"

with open ("demo.txt","w") as file:
  file.writelines(content)
```
