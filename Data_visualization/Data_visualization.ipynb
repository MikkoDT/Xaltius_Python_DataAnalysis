```python
# to import Matplotlib
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
year = [2020,2021,2022,2023,2024]
total_population = [8952000,8967000,8988760,9091000,9290000]

plt.plot(year,total_population)
plt.title("Population over years")
plt.xlabel('year')
plt.ylabel('total_population')
plt.show()
```

```python
# Example
price = [30,32,20,28,50]
people = [100,115,112,75,82]
plt.scatter(price,people)

plt.title("People and Food Graph ")
plt.xlabel('price')
plt.ylabel('people')
plt.colorbar()
plt.show()
```

```python
# Subplot
x = np.array([0,1,2,6])
y = np.array([3,8,1,9])
plt.subplot(1,2,1) # one row, two, columns, no. of figure
plt.plot(x,y,marker='D',ms=10)
```

```python
# Seaborn
url = "http://bit.ly/wkspdata"
data = pd.read_csv(url)
data.head()
```

```python
data.describe()
```

```python
data.shape
```

```python
data.info()
```

```python
# map the data
data["alive"]=data.survived.map({0:"No",1:"Yes"})
data["class"]=data.pclass.map({1:"First",2:"Second",3:"Third"})
data.head()
```

```python
# Factorplot - used to draw categorical plot
#sns.load_dataset("data")
sns.catplot(x="sex",data=data,kind="count")
```

```python
sns.catplot(x="class",data=data,hue="sex",kind="count")
```

```python
#Age by class
sns.catplot(x="class",y="age",hue="survived",data=data,aspect=2,kind="box")
```

```python
corr = data.corr()

#Generate color, map, red, blue
colorsintheheatmap = sns.diverging_palette(220,10,as_cmap=True)
sns.heatmap(corr,cmap=colorsintheheatmap,annot=True,fmt=".2f")
```
