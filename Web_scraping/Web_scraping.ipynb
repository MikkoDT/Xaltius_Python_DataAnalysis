```python
# Web scraping: To extract the data from web

from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from webdriver_manager.chrome import ChromeDriverManager

#step1: 
driver = webdriver.Chrome(ChromeDriverManager().install())
driver.maximize_window()
driver.get("https://www.python.org/")
assert "Python" in driver.title

#step2:
elem = driver.find_element(by='name',value="q")
elem.clear()
elem.send_keys("Applications for Python")

#step3:
elem.send_keys(Keys.RETURN)

#step4:
assert "No result found." not in driver.page_source

#step5:
ul = driver.find_element(by='class name',value='list-recent-events.menu')
li_s = ul.find_elements(by='tag name',value='li')

#step6:
print(li_s[1].text)

#step7:
driver.close()
```
